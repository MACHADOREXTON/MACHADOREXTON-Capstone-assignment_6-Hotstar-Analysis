```python
# Fill missing values in 'Running_time' with the median of the column
df['Running_time'] = df['Running_time'].fillna(df['Running_time'].median())
```

```python
# Fill missing values in 'Seasons' and 'Episodes' with 0
df['Seasons'] = df['Seasons'].fillna(0)
df['Episodes'] = df['Episodes'].fillna(0)
```
